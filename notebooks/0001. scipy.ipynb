# Scipy

```python
%matplotlib inline
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from matplotlib.pyplot import imshow
from scipy.signal import convolve2d
```

## PDF and CDF

```python
x = np.linspace(-6, 6, 1_000)
x.shape
```

```python
fx = norm.pdf(x, loc=0, scale=1)
fx.shape
```

```python
plt.plot(x, fx)
plt.title('PDF of standard normal')
```

```python
Fx = norm.cdf(x, loc=0, scale=1)
Fx.shape
```

```python
plt.plot(x, Fx)
plt.title('CDF of standard normal')
```

```python
logfx = norm.logpdf(x, loc=0, scale=1)
logfx.shape
```

```python
# log pdf of the standard normal
plt.plot(x, logfx)
plt.title('Log PDF of standard normal')
```

## Convolution

```python
im = Image.open('data/tux.svg.png')
imshow(np.asarray(im))
```

```python
im_gray = np.mean(im, axis=2)
imshow(np.asarray(im_gray))
```

```python
x = np.linspace(-6, 6, 50)
fx = norm.pdf(x, loc=0, scale=1)
```

```python
# gaussian filter
filt = np.outer(fx, fx)
```

```python
plt.imshow(filt, cmap='gray')
```

```python
out = convolve2d(im_gray, filt)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(im_gray, cmap='gray')
plt.subplot(1, 2, 2)
plt.imshow(out, cmap='gray')
```
